```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
pd.options.display.max_rows = 99
pd.options.display.max_columns = 999
from __future__ import division
from datetime import datetime,timedelta
```

```python
# def to_str(x): #used below to pad hours with 0 if less than 10
#     if x < 10:
#         return '0' + x.astype(str)
#     else:
#         return x.astype(str)
    
def pad(x):
    time = x.astype(str).zfill(4)
    return time[:2] + ':' + time[2:]
```

```python
#contains headers for all data but I edited "nothing"
df_header = pd.read_csv('data/sensor_data/colorado6months/header.csv')
headers = df_header.columns
df_sensor = pd.read_csv('data/sensor_data/colorado6months/20140401.csv',header=None) #sensors
df_sensor.columns = headers
df_sensor['MST'] = df_sensor['MST'].map(pad)
#make a datetime index:
df_sensor['datetime'] = pd.to_datetime(df_sensor['Year'].astype(str)+
                                       df_sensor['DOY'].astype(str)+
                                       df_sensor['MST'].astype(str), 
                                       format='%Y%j%H:%M')
#convert to utc time, website shows that they disregard daylight savings time
df_sensor['datetime'] = df_sensor['datetime'] + pd.Timedelta(hours=7) 
df_sensor.set_index(['datetime'],inplace=True)
df_sensor = df_sensor.resample('H')
df_sensor                                 
                   
```

```python
df_output = pd.read_csv('data/pvoutput/pvoutput6months/20140401.csv') #pvoutput
#df_output['datetime'] = df_output['datetime'] + pd.Timedelta(hours=6) #convert to utc time
df_output['datetime'] = df_output['datetime'].apply(pd.to_datetime)
df_output['datetime'] = df_output['datetime'] + pd.Timedelta(hours=6) #convert to utc time
df_output.set_index(['datetime'],inplace=True)
df_output = df_output.resample('H')
df_output.drop('Unnamed: 12',axis=1,inplace=True)
df_output
```
